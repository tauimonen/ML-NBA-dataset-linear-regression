# üèÄ NBA Team Traditional Stats Dataset (2023‚Äì24 Season for training the model and 2024-25 for the testing)

Datasets contain per-game team statistics for all NBA teams during the seasons.  
Each row represents one team, including both raw performance metrics and league-wide ranking values for each statistic.

***This analysis focuses only on actual NBA teams (30 teams)***

---

## üìã Dataset Overview

| Property | Description |
|-----------|-------------|
| **Observations (rows)** | 30 teams |
| **Columns (features)** | 54 |
| **Data Types** | `int64`, `float64`, `object` |
| **Source** | [NBA Stats API - LeagueDashTeamStats](https://stats.nba.com/stats/leaguedashteamstats) |

---

## Variable Descriptions

### Basic Team Information
| Column | Description | Type |
|---------|--------------|------|
| `TEAM_ID` | Unique numeric identifier for each team. | `int64` |
| `TEAM_NAME` | Official team name (e.g., *Boston Celtics*, *Golden State Warriors*). | `object` |

---

### Team Performance Summary
| Column | Description | Type |
|---------|--------------|------|
| `GP` | Games played. | `int64` |
| `W` | Wins. | `int64` |
| `L` | Losses. | `int64` |
| `W_PCT` | Win percentage (`W / GP`). | `float64` |
| `MIN` | Average minutes played per game. | `float64` |

---

### Shooting Statistics
| Column | Description | Type |
|---------|--------------|------|
| `FGM` | Field goals made per game. | `float64` |
| `FGA` | Field goals attempted per game. | `float64` |
| `FG_PCT` | Field goal percentage. | `float64` |
| `FG3M` | 3-point field goals made per game. | `float64` |
| `FG3A` | 3-point field goals attempted per game. | `float64` |
| `FG3_PCT` | 3-point field goal percentage. | `float64` |
| `FTM` | Free throws made per game. | `float64` |
| `FTA` | Free throws attempted per game. | `float64` |
| `FT_PCT` | Free throw percentage. | `float64` |

---

### Rebounding & Passing
| Column | Description | Type |
|---------|--------------|------|
| `OREB` | Offensive rebounds per game. | `float64` |
| `DREB` | Defensive rebounds per game. | `float64` |
| `REB` | Total rebounds per game. | `float64` |
| `AST` | Assists per game. | `float64` |

---

### Defense & Possession
| Column | Description | Type |
|---------|--------------|------|
| `TOV` | Turnovers per game. | `float64` |
| `STL` | Steals per game. | `float64` |
| `BLK` | Blocks per game. | `float64` |
| `BLKA` | Shots blocked against (times the teams shots were blocked). | `float64` |
| `PF` | Personal fouls committed per game. | `float64` |
| `PFD` | Personal fouls drawn per game. | `float64` |
| `PTS` | Points scored per game. | `float64` |
| `PLUS_MINUS` | Average point differential (team points minus opponent points). | `float64` |

---

### Ranking Metrics
Each main statistic has a corresponding **league ranking** column (1 = best).  
For example, `FG3A_RANK = 1` means the team attempted the most 3-point shots in the league.

| Column Pattern | Description | Example |
|----------------|-------------|----------|
| `*_RANK` | League rank for the corresponding stat. | `PTS_RANK`, `REB_RANK`, `FG3A_RANK`, etc. |

---

## Notes
- All statistics are **per-game averages** unless otherwise stated.  
- Rankings are relative to all teams in the dataset.  
- The dataset may contain more than 30 entries (e.g., historical or combined team records).  
- Data sourced directly from NBA Stats official API.

```python
from nba_api.stats.endpoints import leaguedashteamstats
import pandas as pd
from sklearn.metrics import root_mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression

# Hae 2023‚Äì24 kauden "traditional team stats" (eli se sama taulukko)
data = leaguedashteamstats.LeagueDashTeamStats(
    season='2023-24',
    season_type_all_star='Regular Season',
    per_mode_detailed='PerGame',  # sama kuin sivulla: "Per Game"
    league_id_nullable='00'   # üîπ vain NBA
)

test_data = leaguedashteamstats.LeagueDashTeamStats(
    season='2024-25',
    season_type_all_star='Regular Season',
    per_mode_detailed='PerGame',  # sama kuin sivulla: "Per Game"
    league_id_nullable='00'   # üîπ vain NBA
)

df = data.get_data_frames()[0]
test_df = test_data.get_data_frames()[0]

# tulostetaan ensimm√§iset rivit
print(df.head())
print(test_df.head())

# tallennetaan CSV-tiedostoiksi
df.to_csv("nba_team_stats_2023_24.csv", index=False)
test_df.to_csv("nba_team_stats_2024_25.csv", index=False)
```

```python
# Lista selitt√§vist√§ muuttujista (NBA-numeraaliset)
features = [
    "GP", "W", "L", "MIN", "FGM", "FGA", "FG_PCT",
    "FG3M", "FG3A", "FG3_PCT", "FTM", "FTA", "FT_PCT",
    "OREB", "DREB", "REB", "AST", "TOV", "STL", "BLK", "BLKA",
    "PF", "PFD", "PTS", "PLUS_MINUS"
]

# M√§√§ritell√§√§n grid 
n_cols = 6
n_rows = (len(features) + n_cols - 1) // n_cols  # py√∂ristet√§√§n yl√∂sp√§in

fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols*4, n_rows*4))
axes = axes.flatten()  # helpottaa indeksointia

for i, col in enumerate(features):
    sns.regplot(
        data=df,
        x=col,
        y="W_PCT",
        ax=axes[i],
        scatter_kws={"alpha":0.6}
    )
    axes[i].set_title(f"{col} vs W_PCT")

for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
